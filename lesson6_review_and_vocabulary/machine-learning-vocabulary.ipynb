## 在今天的课程中将涉及的新词汇：



ground truth: the ground truth is the correct or actual output of a dataset. 真实值是数据集的正确或实际输出。

function: a function is a block of code that performs a specific task. 函数是一段代码块，它完成特定的任务。

empirical: 经验的

objective function: the objective function is the function that the machine learning model tries to minimize or maximize. 目标函数是机器学习模型试图最小化或最大化的函数。

loss function: the loss function is the function that measures the error between the predicted output and the actual output. 损失函数是衡量预测输出和实际输出之间的误差的函数。

gradient descent: gradient descent is an optimization algorithm that is used to minimize the loss function. 梯度下降是一种优化算法，用于最小化损失函数。

epoch: an epoch is a single iteration of the training process. 一个epoch是训练过程中的一次迭代。

batch size: the batch size is the number of training examples that are processed in each iteration. 批处理大小是每次迭代中处理的训练样本数。

regularize: to prevent overfitting. 正则化是为了防止过拟合。

hyperparameter: a hyperparameter is a parameter that is set before training a model. 超参数是模型训练之前设置的参数。

parameter: a parameter is a variable that is used to control the behavior of a model. 参数是模型行为的变量。

regression: a regression model is a model that predicts a continuous variable. 回归模型是一种预测连续变量的模型。

注：连续的反义词是 discrete 离散。
    比如1,2,3,4,5是离散的，而1~5是连续的，你可以取任意大于等于1小于等于5的小数。

linear regression: a type of regression model that assumes a linear relationship between the input and output variables. 线性回归是一种回归模型，它假设输入和输出变量之间存在线性关系。

Logistic regression: a type of regression model that is used for binary classification problems. 逻辑回归是一种回归模型，它用于二分类问题。
